```python
dfTiONiO['ClassType'] = ""

dfTiONiO.loc[dfTiONiO['PCE'] >= 11, 'ClassType'] = 'Good'
dfTiONiO.loc[(dfTiONiO['PCE']<11) & (dfTiONiO['PCE']>=5), 'ClassType'] = 'Okay'
dfTiONiO.loc[dfTiONiO['PCE']<5, 'ClassType'] = 'Poor'

dfTiONiO.sort_values('PCE', ascending=False)

```

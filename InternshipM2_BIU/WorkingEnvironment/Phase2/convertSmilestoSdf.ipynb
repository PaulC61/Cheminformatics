```python
import pandas as pd
from rdkit.Chem import PandasTools
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.ipython_useSVG=True
```

```python
metalsMixes = pd.read_csv('metalsMixes.csv', index_col=0)
noMetalsMixes = pd.read_csv('noMetalsMixes.csv', index_col=0)
metalsNoMixes = pd.read_csv('metalsNoMixes.csv', index_col=0)
noMetalsNoMixes = pd.read_csv('noMetalsNoMixes.csv', index_col=0)

metalsMixes
```

```python
PandasTools.AddMoleculeColumnToFrame(metalsMixes, 'Molecule SMILE', 'MolCol', includeFingerprints=False)
PandasTools.AddMoleculeColumnToFrame(noMetalsMixes, 'Molecule SMILE', 'MolCol', includeFingerprints=False)
PandasTools.AddMoleculeColumnToFrame(metalsNoMixes, 'Molecule SMILE', 'MolCol', includeFingerprints=False)
PandasTools.AddMoleculeColumnToFrame(noMetalsNoMixes, 'Molecule SMILE', 'MolCol', includeFingerprints=False)

metalsMixes
```

```python
PandasTools.WriteSDF(metalsMixes, 'metalsMixes.sdf', molColName='MolCol', idName='Dragon Name')
PandasTools.WriteSDF(noMetalsMixes, 'noMetalsMixes.sdf', molColName='MolCol', idName='Dragon Name')
PandasTools.WriteSDF(metalsNoMixes, 'metalsNoMixes.sdf', molColName='MolCol', idName='Dragon Name')
PandasTools.WriteSDF(noMetalsNoMixes, 'noMetalsNoMixes.sdf', molColName='MolCol', idName='Dragon Name')
```

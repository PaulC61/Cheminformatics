```python
import pandas as pd
import numpy as np

rawGlobalDyes = pd.read_csv('../../rawDSSCDBdata.csv')

rawGlobalDyes.head()
```

```python
rawGlobalDyes['Molecule keywords'] = rawGlobalDyes['Molecule keywords'].str.lower()

rawGlobalDyes.groupby(['Molecule keywords']).count().sort_values('PCE', ascending=False)

rawGlobalDyes
```

```python
rawGlobalDyes

# solar simulator settings

varSolSim = ["AM 1.5G 100 mW/cm2", 
             "AM 1.5",
             "AM 1.5 100 mW/cm2", 
             "AM 1.5 100mW/cm2", 
             "AM 1.5G 100mW/cm2",
             "AM 1.5G, 100 mW/cm2", 
             "AM 1.5G 100 mW/cm2", 
             "AM 1.5G, 100mW/cm2",
             "AM 1.5G", 
             "AM 1.5 100mW/cm2", 
             "AM 1.5G, 100 mw/cm2", 
             "AM 1.5 G",
             "AM1.5G, 100 mW/cm2", 
             "AM1.5 G 100 mW/cm2" 
             ,"AM1.5G"
            ]

for index in range(len(varSolSim)):
    rawGlobalDyes.loc[(rawGlobalDyes['Solar simulator']==varSolSim[index]), 'Solar simulator'] = "AM 1.5G 100mW/cm2"

dyesFilterSolarSim = rawGlobalDyes.loc[rawGlobalDyes['Solar simulator'].str.contains('100')]

dyesFilterSolarSim
```

```python
varElectrolyte = ["iod", "Iod","LiI", "I2", "AN-I", "Mosalyte", "HI-30"]

for index in range(len(varElectrolyte)):
    dyesFilterSolarSim.loc[(dyesFilterSolarSim['Electrolyte'].str.contains(varElectrolyte[index])), 'Electrolyte'] = "Iodide_Triiodide"

dyesFSolarSimNdElec = dyesFilterSolarSim.loc[dyesFilterSolarSim['Electrolyte']=="Iodide_Triiodide"]

dyesFSolarSimNdElec
```

```python
dyesFCoObjects = dyesFSolarSimNdElec.loc[(dyesFSolarSimNdElec['Co-adsorbent']=="-") & (dyesFSolarSimNdElec['Co-sensitizer']=="-")]

dyesFCoObjects
```

```python
dyesFSemiConductor = dyesFCoObjects.loc[dyesFCoObjects['Semiconductor'].str.contains('TiO2')]

dyesFSemiConductor
```

```python
dyesFSolarSim = dyesFSemiConductor.loc[dyesFSemiConductor['Solar simulator'].str.contains('100')]

dyesFSolarSim#.loc[dyesFSolarSim['Molecule SMILE'].str.contains('\.', regex=True)]
```

```python
dyesFMixtures = dyesFSolarSim.loc[(~dyesFSolarSim['Molecule SMILE'].str.contains('\.', regex=True))]

dyesFMixtures
```

```python
dyesDropDuplicatesMixtures = dyesFSolarSim.sort_values('PCE', ascending=False).drop_duplicates('Molecule SMILE').sort_index()

dyesDropDuplicatesSingles = dyesFMixtures.sort_values('PCE', ascending=False).drop_duplicates('Molecule SMILE').sort_index()

```

```python
metalsMixes = dyesDropDuplicatesMixtures

metalsMixes.shape
```

```python
metalsNoMixes = dyesDropDuplicatesSingles

metalsNoMixes.shape
```

```python
noMetalsMixes = metalsMixes.loc[(~metalsMixes['Molecule keywords'].str.contains('ruthenium'))
                                & (~metalsMixes['Molecule keywords'].str.contains('zinc'))
                                   & (~metalsMixes['Molecule SMILE'].str.contains("se"))]

noMetalsMixes.shape #1678 -> 1659
```

```python
noMetalsNoMixes = metalsNoMixes.loc[(~metalsNoMixes['Molecule keywords'].str.contains('ruthenium'))
                                & (~metalsNoMixes['Molecule keywords'].str.contains('zinc'))
                                   & (~metalsNoMixes['Molecule SMILE'].str.contains("se"))]

noMetalsNoMixes.shape #1647 -> 1631
```

```python
noMetalsNoMixes.groupby(['Molecule keywords']).mean().sort_values('PCE', ascending=False)
```

```python
noMetalsMixes.columns
```

```python
metalsMixes['Dragon Name'] = metalsMixes['Molecule keywords'] + ' ' + metalsMixes['PCE'].astype(str)
noMetalsMixes['Dragon Name'] = noMetalsMixes['Molecule keywords'] + ' ' + noMetalsMixes['PCE'].astype(str)
metalsNoMixes['Dragon Name'] = metalsNoMixes['Molecule keywords'] + ' ' + metalsNoMixes['PCE'].astype(str)
noMetalsNoMixes['Dragon Name'] = noMetalsNoMixes['Molecule keywords'] + ' ' + noMetalsNoMixes['PCE'].astype(str)
```

```python
metalsMixes.to_csv('metalsMixes.csv')
noMetalsMixes.to_csv('noMetalsMixes.csv')
metalsNoMixes.to_csv('metalsNoMixes.csv')
noMetalsNoMixes.to_csv('noMetalsNoMixes.csv')
```

```python
metalsMixesSMI=pd.concat([metalsMixes['Dragon Name'], metalsMixes['Molecule SMILE']], axis=1)
noMetalsMixesSMI=pd.concat([noMetalsMixes['Dragon Name'], noMetalsMixes['Molecule SMILE']], axis=1)
metalsNoMixesSMI=pd.concat([metalsNoMixes['Dragon Name'], metalsNoMixes['Molecule SMILE']], axis=1)
noMetalsNoMixesSMI=pd.concat([noMetalsNoMixes['Dragon Name'], noMetalsNoMixes['Molecule SMILE']], axis=1)


```

```python
metalsMixesSMI.to_csv('metalsMixes.smi', sep='\t', index=False, header=False)
noMetalsMixesSMI.to_csv('noMetalsMixes.smi', sep='\t', index=False, header=False)
metalsNoMixesSMI.to_csv('metalsNoMixes.smi', sep='\t', index=False, header=False)
noMetalsNoMixesSMI.to_csv('noMetalsNoMixes.smi', sep='\t', index=False, header=False)
```

```python
# for 3D Canvas Descriptors
cols = list(qikPharmDataset.columns)
X = qikPharmDataset[cols[:251]]
Y = qikPharmDataset['PCE']

# for PaDEL descriptors
# X = descriptors.drop('Name', axis=1)
# Y = dyeData['PCE']
X
```

```python
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X
```

```python
X = X.dropna(axis=1, how='any')

X
```

```python

from sklearn.feature_selection import VarianceThreshold

def remove_low_variance(input_data, threshold=0.1):
    selection = VarianceThreshold(threshold)
    selection.fit(input_data)
    return input_data[input_data.columns[selection.get_support(indices=True)]]

X = remove_low_variance(X, threshold=0.1)
X
```

```python
# Split into Train and Test
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25,random_state=42)

X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

modelRF = RandomForestRegressor()
modelSVR = SVR()
modelGB = GradientBoostingRegressor()
modelKNN = KNeighborsRegressor()

modelRF.fit(X_train, Y_train)
modelSVR.fit(X_train, Y_train)
modelGB.fit(X_train, Y_train)
modelKNN.fit(X_train, Y_train)

```

```python
X_test.to_csv('testSet1D2D.csv')
```

```python
X_test = X_test.dropna()
Y_test = Y_test.drop(924)
```

```python
Y_train_pred_RF = modelRF.predict(X_train)
Y_train_pred_SVR = modelSVR.predict(X_train)
Y_train_pred_GB = modelGB.predict(X_train)
Y_train_pred_KNN = modelKNN.predict(X_train)

Y_test_pred_RF = modelRF.predict(X_test)
Y_test_pred_SVR = modelSVR.predict(X_test)
Y_test_pred_GB = modelGB.predict(X_test)
Y_test_pred_KNN = modelKNN.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

mae_train_RF = mean_absolute_error(Y_train, Y_train_pred_RF)
mae_train_SVR = mean_absolute_error(Y_train, Y_train_pred_SVR)
mae_train_GB = mean_absolute_error(Y_train, Y_train_pred_GB)
mae_train_KNN = mean_absolute_error(Y_train, Y_train_pred_KNN)

mae_test_RF = mean_absolute_error(Y_test, Y_test_pred_RF)
mae_test_SVR = mean_absolute_error(Y_test, Y_test_pred_SVR)
mae_test_GB = mean_absolute_error(Y_test, Y_test_pred_GB)
mae_test_KNN = mean_absolute_error(Y_test, Y_test_pred_KNN)

r2_train_RF = r2_score(Y_train, Y_train_pred_RF)
r2_train_SVR = r2_score(Y_train, Y_train_pred_SVR)
r2_train_GB = r2_score(Y_train, Y_train_pred_GB)
r2_train_KNN = r2_score(Y_train, Y_train_pred_KNN)

r2_test_RF = r2_score(Y_test, Y_test_pred_RF)
r2_test_SVR = r2_score(Y_test, Y_test_pred_SVR)
r2_test_GB = r2_score(Y_test, Y_test_pred_GB)
r2_test_KNN = r2_score(Y_test, Y_test_pred_KNN)

(mae_train_RF, mae_test_RF, r2_train_RF, r2_test_RF), (mae_train_SVR, mae_test_SVR, r2_train_SVR, r2_test_SVR), (mae_train_GB, mae_test_GB, r2_train_GB, r2_test_GB), (mae_train_KNN, mae_test_KNN, r2_train_KNN, r2_test_KNN)
```
